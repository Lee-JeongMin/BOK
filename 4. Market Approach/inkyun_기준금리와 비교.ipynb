```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

rate = pd.read_csv('c:/users/user/documents/base_rate_preprocessing.csv')
df = pd.read_csv('c:/users/user/Tone_doc.csv')
```

```python
tone_rate = df.merge(rate, on='date', how='outer').dropna().reset_index(drop=True)
```

```python
sns.heatmap(tone_rate.corr(), annot = True, fmt='.2f', linewidths=.5, cmap = 'Blues')
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax2 = ax1.twinx()

line1 = ax1.plot(tone_rate.date, tone_rate.TONE_Doc, color='orange', label = 'tone_market')
line2 = ax2.plot(tone_rate.date, tone_rate.rate, color='green', label = 'base_rate')

lines = line1 + line2
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc=1)

plt.xticks(tone_rate.date[::12])
```
