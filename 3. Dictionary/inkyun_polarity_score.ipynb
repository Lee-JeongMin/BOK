```python
import pandas as pd
from tqdm.notebook import tqdm
total_df = pd.read_csv('c:/Users/user/downloads/total_up_down.csv')

total_df['polar_score'] = 0
total_df['polarity'] = 0
total_df
```

```python
total_cnt = sum(total_df.total_count)
up_cnt = sum(total_df.up_count)
down_cnt = sum(total_df.down_count)
```

```python
p_hawkish = up_cnt/total_cnt
p_dovish = down_cnt/total_cnt

for row in tqdm(range(total_df.shape[0])):
    total_df.loc[row,'polar_score'] = ((total_df.loc[row,'up_count']+0.001/total_df.loc[row,'total_count'])/p_hawkish)/((total_df.loc[row,'down_count']+0.001/total_df.loc[row,'total_count'])/p_dovish)
    if total_df.loc[row,'polar_score'] > (1.3/1):
        total_df.loc[row,'polarity'] = 'Hawkish'
    elif total_df.loc[row,'polar_score'] < (1/1.3):
        total_df.loc[row,'polarity'] = 'Dovish'
    else:
        total_df.loc[row,'polarity'] = 'nothing'
```

```python
total_df.polarity.value_counts()
```

```python
hawk_df = total_df[total_df['polarity'] == 'Hawkish']
dove_df = total_df[total_df['polarity'] == 'Dovish']
```
