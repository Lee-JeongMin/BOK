```python
import pandas as pd
import re
import numpy as np
```

```python
merge = pd.read_csv('c:/Users/user/downloads/news_report_mpb.csv')
```

```python
# ngrams 추출하기2 - split사용
doc_dict = {}

total_ngram_list = []
pos_ngram_list = []
neg_ngram_list = []
same_count = 0
for idx1, ngrams in enumerate(merge.ngram):
    doc_dict[idx1] = []
    ngram_list = []
    label = merge.label[idx1]
    if label == 'up':
        pos_ngram_list.append(merge.ngram[idx1].replace('[','').replace(']','').replace('\'','').replace(' ','').split(','))
    elif label == 'down':
        neg_ngram_list.append(merge.ngram[idx1].replace('[','').replace(']','').replace('\'','').replace(' ','').split(','))
        
    for idx2, ngram in enumerate(ngrams.split(',')):
        temp = ngram.replace('\'','').replace(' ','').replace('[','').replace(']','')
        if temp == '':
            continue
        ngram_list.append(temp)
        total_ngram_list.append(temp)
    doc_dict[idx1] = ngram_list
```

```python
print('긍정, 부정문서의 수:')
print(len(pos_ngram_list),len(neg_ngram_list)) # same 라벨까지 구하여 더하면 딱 4715개가 나온다. 3584+1131
```

```python
total_ngram_set = set(total_ngram_list)
df_count = pd.DataFrame({'ngram':list(total_ngram_set),'total_count':0,'up_count':0,'down_count':0,'polar_score':0,'polarity':0})
df_count
```

```python
# 1. 문서에 1번 이상 등장하면 1로 카운트

pos_dict = {}
neg_dict = {}
print('전체 문서 수: ',len(total_ngram_set))
for idx, ngram in enumerate(list(total_ngram_set)):
    if idx % 100 == 0:
        print(idx,'번째 문서 작업 중')
    pos_count = 0
    neg_count = 0
    for pos_ngram in pos_ngram_list:
        if ngram in pos_ngram:
            pos_count = pos_ngram.count(ngram)
        pos_dict[ngram]=pos_count
    
    for neg_ngram in neg_ngram_list:
        if ngram in neg_ngram:
            neg_count = neg_ngram.count(ngram)
        neg_dict[ngram]=pos_count
        break
    break
```
